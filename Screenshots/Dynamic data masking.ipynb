# DYNAMIC DATA MASKING

```python
%sql
select * from Katalog.sparksql.ecome_ext
```

```python
%sql
create or replace function Katalog.sparksql.dynamic_mask(Mask_user_id String)
Return
Case  
    when is_account_group_member('main') then Mask_user_id else '****-**' end
```

```python
%sql
alter table Katalog.sparksql.ecome_ext
alter column user_id set mask katalog.sparksql.dynamic_mask
    
```
