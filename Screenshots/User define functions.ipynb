# Functions 

### User define functions | scalar function


```python

%sql
create or replace function Katalog.sparksql.discounted_price(P_price decimal(10,2))
returns decimal(10,2)
language SQL
return P_price * 0.90
```

```python
%sql
select Price_per_unit, Katalog.sparksql.discounted_price(Price_per_unit) as UDF from katalog.sparksql.ecome_ext
```

### User define Table functions | Table function

```python
%sql
create or replace function katalog.sparksql.category(C_category String)
returns table 
language SQL
return
(
select * from katalog.sparksql.ecome_ext where C_category = product_category
)
    
```

```python
%sql
select * from katalog.sparksql.category('Electronics')  -- just like where condition we cna pass the specific category and executes the table
```
