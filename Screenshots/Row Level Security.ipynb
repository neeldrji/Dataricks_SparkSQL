# ROW LEVEL SECURITY

### Mapping Table

```python
%sql
create table katalog.sparksql.mapping_table
(
    Payment_category String,
    Email String
)
```

```python
%sql
insert into katalog.sparksql.mapping_table
values
('Credit Card', 'manuship58@gmail.com'),
('Debit Card', 'manuship58@gmail.com'),
('PayPal', 'newuser@gmail.com'),
('UPI', 'newuser@gmail.com')
```

```python
%sql
select * from katalog.sparksql.mapping_table
```

```python
%sql
select * from katalog.sparksql.mapping_table
where email = current_user()
AND Payment_category = 'Debit Card'
```

## Converting mapping table in Boolean Function

```python
%sql
select exists
(select * from katalog.sparksql.mapping_table
where email = current_user()
AND Payment_category = 'Debit Card'
)
```

```python
%sql
Create or replace function  katalog.sparksql.Row_level_security(P_Payment_mathod STRING)
RETURNS boolean
LANGUAGE SQL
RETURN
(
  EXISTS
 (select * from katalog.sparksql.mapping_table
  where email = current_user()
  AND Payment_category = P_Payment_mathod
 )
)
```

```python
%sql
alter table katalog.sparksql.ecome_ext
set row Filter katalog.sparksql.Row_level_security on (Payment_method)
```

```python
%sql
select * from katalog.sparksql.ecome_ext
```
